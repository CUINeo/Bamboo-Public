```python
import re
import pandas as pd
import numpy as np
from abort_analyze_helper import *
```

```python
match_get_row = re.compile(r"txn=(\d*) row=(\d*) type=(\d*)")
match_status = re.compile(r"txn=(\d*) status=(\d*)")
```

```python
df = pd.DataFrame()
```

```python
f = open("debug.txt")
```

```python
last_row = ""
ts = 0
for line in f:
    if "txn=" not in line:
        continue
    line = line.strip()
    if "row=" in line:
        matched = match_get_row.search(line)
        txn, row, access_t = matched.groups(1)
        txn = int(txn)
        access_t = LOCK_T[int(access_t)]
        if (row in df.index) and (isinstance(df.loc[row,ts], str)):
            ts += 1
        df.loc[row, ts] = "[%d] %d, %s"%(ts, txn, access_t)
    else:
        matched = match_status.search(line)
        txn, status = matched.groups(1)
        status = STATUS[int(status)]
        print("[%d] txn-%s, %s"%(ts, txn, status))
```

```python
def print_locktable(ts):
    for rid in range(df.shape[0]):
        locktable = df.iloc[rid, :ts][df.iloc[rid, :ts].notnull()]
        if len(locktable) > 1:
            print(locktable.values)
```
