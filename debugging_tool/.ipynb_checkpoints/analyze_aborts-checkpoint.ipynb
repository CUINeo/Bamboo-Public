```python
import re
import pandas as pd
import numpy as np
from abort_analyze_helper import *
```

```python
match_get_row = re.compile(r"txn=(\d*) row=(\d*) type=(\d*)")
match_status = re.compile(r"txn=(\d*) status=(\d*)")
match_wound = re.compile(r"txn=(\d*) \((\d*),ts=(\d*)\) want to wound txn=(\d*) \((\d*),ts=(\d*)\) on row=(\d*)")
```

```python
def print_locktable(ts):
    for rid in range(df.shape[0]):
        locktable = df.iloc[rid, :ts][df.iloc[rid, :ts].notnull()]
        if len(locktable) > 1:
            print(locktable.values)
```

```python
df = pd.DataFrame()
```

```python
f = open("debug.txt")
```

```python
last_row = ""
ts = 0
wound_set = set()
cascading = 0
wound = 0
committed = 0
for line in f:
    if "txn=" not in line:
        continue
    line = line.strip()
    if "ts=" in line:
        matched = match_wound.search(line)
        wounder, type1, ts1, wounded, type2, ts2, row = matched.groups(1)
        if (row in df.index) and (isinstance(df.loc[row,ts], str)):
            ts += 1
        df.loc[row, ts] = "[%d] %d (%s, %s) wounded %d (%s, %s)"%(
            ts, int(wounder), LOCK_T[int(type1)], ts1, 
            int(wounded), LOCK_T[int(type2)], ts2)
        print(df.loc[row, ts])
        wound_set.add(int(wounded))
        wound += 1
    elif "row=" in line:
        matched = match_get_row.search(line)
        txn, row, access_t = matched.groups(1)
        txn = int(txn)
        access_t = ACCESS_T[int(access_t)]
        if (row in df.index) and (isinstance(df.loc[row,ts], str)):
            ts += 1
        df.loc[row, ts] = "[%d] %d, %s"%(ts, txn, access_t)
    else:
        matched = match_status.search(line)
        txn, status = matched.groups(1)
        status = STATUS[int(status)]
        print("[%d] txn-%s, %s"%(ts, txn, status))
        txn = int(txn)
        if status == "aborted":
            if txn in wound_set:
                wound_set.remove(txn)
            else:
                cascading+=1
        else:
            committed += 1
```
